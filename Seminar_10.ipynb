bill_length_mm - длина клюва в мм;
bill_depth_mm - толщина клюва в мм;
flipper_length_mm - длина крыла в мм;
body_mass_g - масса тела в граммах;
sex - пол особи.

```python
penguins = sns.load_dataset("penguins")
penguins.head(10)
```

```python
penguins.info()
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue='species', size='island')
```

```python
sns.scatterplot(data=penguins, x="island", y="body_mass_g", hue='species', size='bill_length_mm')
```

```python
lst = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
g = sns.PairGrid(penguins[lst])
g.map(sns.scatterplot)
```

```python
data = penguins.pivot_table(index='species',columns='island',values='body_mass_g')
sns.heatmap(data)
```
